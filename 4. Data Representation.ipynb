```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

<h1>Discrete Variables:</h1>

```python
y=np.unique(np.random.randint(0,4,100))
plt.hist(y)
```

<h1>Continuous Variables:</h1>

```python
import scipy.stats as stats
nu = 20
sigma = 2
h = sorted(np.random.normal(nu,sigma,100))
abs(20 - np.mean(h))<2
```

```python
plt.hist(h, 30)
```

```python
import scipy.stats as stats
```

```python
plt.figure(figsize=(10,5))
fit = stats.norm.pdf(h, np.mean(h), np.std(h))
plt.plot(h,fit,'-o')
plt.hist(h, density=True)
```
